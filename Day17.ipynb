# Getting dataset from an API
    ->here, we requests the API to borrow us the data

```python
import pandas as pd
import requests
```

```python
#we will use requests.get() fuction to get the data from the url and then we will use pandas to read the data and then we will use pandas to analyze the data
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1')

#here, behind the scene the requests librry requests the API to give us the data and then it gives us the data in the form of a response object 
#and we can use the .json() method to get the data in the form of a dictionary 

#we get the data in the form of a dictionary and then we can use the pandas library to read the data and then we can use the pandas library to analyze the data and convert it to dataframe.
#also we are storing the thing that iscoming as per our request in a variable called response 
```

whenever we hits for some requests there are some conditional codes of the HTTPS protocol

1] 200 -> every thing went perfect

2] 404 -> the thing for which we are hitting is currently UNAVAILABLE

3] 500 -> server down

```python
#we are now going to call response.json() to get the data explicitly in json format
response.json()
```

```python
#if we wanted a particular thing then we call it like the column name or key name.
response.json()['results']
```

```python
#to convert this datasetwe use pd.DataFrame(response.json()['results']) as the syntax where the res.json() is for giving the data in json format 
#and then we are passing the required portion which is results into the pd.DataFrame() function to convert it to dataframe.

pd.DataFrame(response.json()['results'])
```

```python
pd.DataFrame(response.json()['results']).head(2)
```

```python
df = pd.DataFrame(response.json()['results'])[['id','title','release_date','overview','popularity','vote_average','vote_count']]
#note that we have to pass the column name in the form of list
```

```python
df
```

this is a part of our dataset. now we have to form a loop which will run the no. of times the total no. of pages are there. we will append the new dataset repetedly

```python
df1 = pd.DataFrame()
```

```python
df1
```

```python
#now we will make a loop which will append data in the iitially empty dataframe df1
for i in range(1,545):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
    #the piece of code written above is saying that i have formatted the url such thatas value of 'i' changes the page no. changes and hence change the url
    df = pd.DataFrame(response.json()['results'])[['id','title','release_date','overview','popularity','vote_average','vote_count']]
    df1 = pd.concat([df1, df], ignore_index=True)
    #in new pandas append is removed so we use the above syntax -- learn it
    #we are using ignore_index = True because we want to ignore the index of the inner dataframe b/z inner df will repetedly have index from 0 to 19 and not doing this will lead to overwriting in the dataframe
```

```python
#now we can save this dataframe to a csv file using the to_csv() method of pandas dataframe
df1.to_csv('top_rated_movies.csv')
```
