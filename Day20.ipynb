```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv(r'D:/ML/DataSet/Titanic_Dataset.csv')
df.sample(5)
```

```python
df.set_index('PassengerId',inplace = True)
```

```python
df.head(5)
```

the process is mainly plotting graphs..

## 1}Categorical data

###             a]Countplot

survived , pclass,sex,sibsp,parch,embarked is a categorical data

```python
df['Survived'].value_counts()
```

```python
sns.countplot(x = 'Survived',data = df)
```

```python
df['Survived'].value_counts().plot(kind = 'bar')
```

```python
sns.countplot(x = 'Embarked', data=df)
```

countplot is a function which will give the graph based on number.

if we wanted in terms of percentage in that case the pie chart of the seaborn library

### b]PieChart

```python
df['Survived'].value_counts().plot(kind = 'pie')
```

if we wanted percentage then we use "autopct = %.2f" inside the plot function

```python
df['Survived'].value_counts().plot(kind = 'pie' , autopct = "%.2f")
```

```python
df['Pclass'].value_counts().plot(kind = 'pie' , autopct = "%.2f")
```

## 2} Numerical Data

### a]Histogram

in this we creates ranges for the numerical dataset

this is a good tool to use in case of the numerical dataset

```python
df['Age'].plot(kind = 'hist',bins = 20) #bins is a parameter used to change the group length in histogram
```

```python
plt.hist(df['Age'],bins=20)     #bins is a parameter used to change the group length in histogram.
```

```python
sns.histplot(df['Age'],bins = 20, kde = True)
```

### b] Distplot
        ->improvement of histogram
        ->this automatically implement kde onto histogram that we have done in above case using parameter
        ->this is kind of PDF(probability density function)
        ->this tells about the probability for a given no. to exist

```python
sns.distplot(df['Age'])     
#this function will be removed in later versions so, learn the above that we have done in last using histogram
```
