### Q6}Are there duplicate values?

### Q7}How is the correlation between Columns?

```python
#correlation = increament or decreament in one variable will cause increament or decreament in other variable.

df.select_dtypes(include = ['number']).corr()

#this is the syntax of the function which will give us the correlation between two variables.
```

```python
#if we waant to get the correlation between only one variable with other all then we do the following thing
df.select_dtypes(include = ['number']).corr()['Survived']
```
