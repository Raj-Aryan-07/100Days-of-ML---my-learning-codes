# Automatic Selection of Value for Imputation

```python
import numpy as np 
import pandas as pd

from sklearn.model_selection import train_test_split,GridSearchCV

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv(r"D:\ML\DataSet\Titanic_Dataset.csv")
df.head()
```

```python
df = df.drop(columns=['PassengerId','Name','Ticket','Cabin'])
```

```python
df.head(3)
```

```python
df = df.rename(columns={'Sex':'Gender'})        #this technique is used to rename the columns
df.head(3)
```

```python
(df.isnull().mean())*100
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=43)
```

```python
X_train.head(3)
```

```python
# our plan is to first impute the age column and then to do scaling using StandardScaler
numerical_features = ['Age', 'Fare']
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_features = ['Embarked', 'Gender']
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ohe',OneHotEncoder(handle_unknown='ignore'))
])

#this piece of code actually allows to apply the required steps when required only like - 
# - when we have all values then imputer will not work.
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)
```

```python
clf = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression())
])
```

```python
from sklearn import set_config

set_config(display='diagram')
clf

#the motive behind this code is to diagramatically show how the work is taking place| this can also be done using simply clf.
```

now, after doing all preprocessing with a common imputation , now we will find the best one using the gridsearch cv

```python
param_grid = {
    'preprocessor__num__imputer__strategy': ['mean', 'median'],
    'preprocessor__cat__imputer__strategy': ['most_frequent', 'constant'],
    'classifier__C': [0.1, 1.0, 10, 100]
}

grid_search = GridSearchCV(clf, param_grid, cv=10)      
#in the grid_search cv we first gave the whole classifier which can be expanded to give the pipeline of the whole work done
#also, we pass a parameter for grid(param_grid) which contains all the parameters for the transformers like all the parameters
# - for the column classifier like num etc. 

#learn the code or remember this
```

```python
grid_search.fit(X_train, y_train)

print(f"Best params:")
print(grid_search.best_params_)     #the best_params_ is a attribute which tells us about the result of all the calculations
# - and tell the best parameters or the strategy that we should use to fill the missing values
```

```python
print(f"Internal CV score: {grid_search.best_score_:.3f}")
```

```python
import pandas as pd

cv_results = pd.DataFrame(grid_search.cv_results_)
cv_results = cv_results.sort_values("mean_test_score", ascending=False)
cv_results[['param_classifier__C','param_preprocessor__cat__imputer__strategy','param_preprocessor__num__imputer__strategy','mean_test_score']]
```
