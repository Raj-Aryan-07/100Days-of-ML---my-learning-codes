## OneHot Encoding with Top Categories

what we do under this topic is that we encode the column having many categories

```python
counts = df['brand'].value_counts()
```

```python
counts
```

```python
df['brand'].nunique()
```

```python
threshold = 100
```

```python
mask = counts<=threshold
print(mask)
```

```python
counts[mask]
```

```python
replace = counts[mask].index
```

```python
pd.get_dummies(df['brand'].replace(replace,'uncommon'),dtype=int).sample(5)

#learn the way to do this to replace the unwanted part under a common name
```
